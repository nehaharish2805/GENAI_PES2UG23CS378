```python
from transformers import pipeline
```

```python
ner_pipeline = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)
```

```python
tweets = [
    "Apple just released a new MacBook Pro today!",
    "I love the design of my new laptop.",
    "Samsung phones have great battery life.",
    "Thinking of buying Apple Care for my device."
]
```

```python
brand_name = "Apple"
```

```python
print(f"Scanning tweets for brand mentions: {brand_name}\n")

for tweet in tweets:
    entities = ner_pipeline(tweet)
    mentioned = False

    # Check NER entities
    for entity in entities:
        if brand_name.lower() in entity["word"].lower():
            mentioned = True

    # Fallback: direct text match
    if brand_name.lower() in tweet.lower():
        mentioned = True

    if mentioned:
        print("MENTION FOUND ✅")
        print("Tweet:", tweet)
        print()
```
